```python
#get averages of player load metrics
player_averages = round(player_load.mean(),2)
player_averages = pd.DataFrame(player_averages)
player_averages
```

```python
def color_corr(value):
  """
  Colors elements in a dateframe
  green if positive and red if
  negative. Does not color NaN
  values.
  """

  if value < .35:
    color = 'black'
  elif value < .50:
    color = 'orange'
  elif value < 0.99:
    color = "green"
  else:
    color = 'gray'

  return 'color: %s' % color
```

```python
corr = player_load.corr().style.applymap(color_corr)
corr
```

```python
corr2 = player_load.corr()
```
